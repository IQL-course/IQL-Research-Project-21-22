```python
import glob
from iso639 import Lang
import csv
import shutil

txtFilters=glob.glob("./../../../data/filtered/alphabets/pud/*.csv")
csvsToAnalyze=glob.glob("./../../../data/non_filtered/corpora/pud/*.csv")

filterLang={}
for txt in txtFilters:
    with open(txt, newline='',encoding="utf-8") as txtOpened:
        pt3Name=txt.split("-")[0].split("\\")[1]
        lines = txtOpened.readlines()
        newFilter=set()
        for line in lines[1:]:
            elems=line.split(",")
            newFilter.add(elems[0].strip("\""))
        if "strokes" in txt:
            pt3Name+="_strokes"
        if "romaji" in txt:
            pt3Name+="_romaji"
        if "pinyin" in txt:
            pt3Name+="_pinyin"
        filterLang[pt3Name]=newFilter
#("jpn" not in csvfile and "zho" not in csvfile and "kor" not in csvfile)
for csvfile in csvsToAnalyze:
    if  True: 
        names=csvfile.split("\\")[1].split("_")[0]
        isoLanIdentifier=names
        isoLan=names
        toAdd=""
        if "strokes" in csvfile:
            toAdd+="_strokes"
        if "romaji" in csvfile:
            toAdd+="_romaji"
        if "pinyin" in csvfile:
            toAdd+="_pinyin"
        with open("./../../../data/filtered/corpora/pud/"+isoLanIdentifier+"_pud"+toAdd+".csv", 'w', encoding='UTF8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(["","word","frequency","n_characters"])
            with open(csvfile, newline='',encoding="utf-8") as csvfileOpened:
                reader = csv.reader(csvfileOpened, delimiter=',', quotechar='"')
                next(reader)
                alphabetSet=filterLang[isoLanIdentifier+toAdd]
                for row in reader:
                    valid=True
                    if "romaji" in csvfile or "pinyin" in csvfile:
                        for letter in row[3]:
                            if letter not in alphabetSet:
                                valid=False
                                break
                        if valid:
                            writer.writerow(row)

                    else:
                        for letter in row[0]:
                            if letter not in alphabetSet:
                                valid=False
                                break
                        if valid:
                            writer.writerow(row)
    else:
        names=csvfile.split("\\")[1].split("_")[0]
        isoLan=names
        isoLanIdentifier=names
        #print(csvfile)
        #toAdd=""
        #if "strokes" in csvfile:
        #    toAdd+="_strokes"
        #if "pinyin" in csvfile:
        #    toAdd+="_pinyin"
        #if "romaji" in csvfile:
        #    toAdd+= "_romaji"
        f_src = open(csvfile, 'rb')
        f_dest = open("./../../../data/filtered/corpora/pud/"+isoLanIdentifier+"_pud.csv", 'wb')
        shutil.copyfileobj(f_src,f_dest) 

```
