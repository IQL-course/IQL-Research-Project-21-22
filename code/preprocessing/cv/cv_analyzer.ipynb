```python
import csv
import glob
from iso639 import Lang

csvsToAnalyze=glob.glob("./common-voice-forced-alignment/*.csv")

with open("../explanatory_tables/common_voice.csv", 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["language", "iso_code", "dialect", "#types", "#tokens", "family"])
    for csvfile in csvsToAnalyze:
        if "phone" not in csvfile: 
            with open(csvfile, newline='',encoding="utf-8") as csvfileOpened:
                reader = csv.reader(csvfileOpened, delimiter=',', quotechar='"')
                next(reader)
                names=csvfile.split("/")[2].split("-")
                isoLan=names[0]
                dialect=""
                if "word" not in names[1]:
                    dialect=names[1]
                totalReps=0
                diffTokens=0
                for row in reader:
                    totalReps+=int(row[0])
                    diffTokens+=1
                writer.writerow([Lang(isoLan).name,isoLan,dialect,diffTokens,totalReps,lang[isoLan]])
```
