```python
import csv
import glob
from iso639 import Lang
import pycountry
from langcodes import *

csvsToAnalyze=glob.glob("./../../../data/cv/*.csv")
with open("./../../../data/descriptive_tables/common_voice.csv", 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["language", "iso_code", "dialect", "#types", "#tokens", "family", "script"])
    for csvfile in csvsToAnalyze:
        if "phone" not in csvfile: 
            with open(csvfile, newline='',encoding="utf-8") as csvfileOpened:
                reader = csv.reader(csvfileOpened, delimiter=',', quotechar='"')
                next(reader)
                names=csvfile.split("/")[5].split("-")
                isoLan=names[0].split("\\")[1]
                dialect=""
                if "word" not in names[1]:
                    dialect=names[1]
                totalReps=0
                diffTokens=0
                for row in reader:
                    totalReps+=int(row[0])
                    diffTokens+=1
                language=pycountry.languages.get(alpha_3=Lang(isoLan).pt3)
                script = scripts[isoLan]
                if((dialect=="HK")|(dialect=="TW")): script = 'Hant'
                print(isoLan,Lang(isoLan).name,lang[isoLan],script)  #test
                if(dialect!=""): print("----------dialect",dialect) #test
                writer.writerow([language.name,Lang(isoLan).pt3,dialect,diffTokens,totalReps,lang[isoLan],script])
```
